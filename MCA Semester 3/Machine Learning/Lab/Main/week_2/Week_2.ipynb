```python
import pandas as pd
import numpy as np

data = {
    'SNO': range(1, 11),
    'REGNO': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'NAME': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'M1': [85, 78, 92, 88, 76, 95, 89, 84, 91, 87],
    'M2': [79, 82, 88, 85, 80, 90, 86, 83, 89, 84],
    'M3': [90, 85, 87, 92, 88, 91, 84, 86, 93, 89],
    'M4': [88, 90, 85, 87, 89, 92, 86, 84, 91, 88],
    'M5': [92, 89, 91, 90, 87, 88, 85, 86, 93, 84]
}

df = pd.DataFrame(data)
print(df)

df.to_csv('students_data.csv', index=False)

```

```python
# Compute statistics
mean = df[['M1', 'M2', 'M3', 'M4', 'M5']].mean()
median = df[['M1', 'M2', 'M3', 'M4', 'M5']].median()
mode = df[['M1', 'M2', 'M3', 'M4', 'M5']].mode().iloc[0]
summary = df[['M1', 'M2', 'M3', 'M4', 'M5']].describe()

print("Mean:\n", mean)
print("Median:\n", median)
print("Mode:\n", mode)
print("5-number summary:\n", summary)
```

```python
import matplotlib.pyplot as plt

# Draw box plot
df[['M1', 'M2', 'M3', 'M4', 'M5']].boxplot()
plt.title("Box Plot of Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# Histograms
df[['M1', 'M2', 'M3', 'M4', 'M5']].hist(bins=10, figsize=(10, 8))
plt.suptitle("Histograms of Marks")
plt.show()

# Bar plots
df[['M1', 'M2', 'M3', 'M4', 'M5']].plot(kind='bar', figsize=(10, 8))
plt.title("Bar Plot of Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# Scatter plots
for subject in ['M1', 'M2', 'M3', 'M4', 'M5']:
    plt.scatter(df['SNO'], df[subject], label=subject)
plt.title("Scatter Plot of Marks")
plt.xlabel("Student Number")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
# Function to find outliers
def find_outliers(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return column[(column < lower_bound) | (column > upper_bound)]

# Display outliers
for subject in ['M1', 'M2', 'M3', 'M4', 'M5']:
    outliers = find_outliers(df[subject])
    print(f"Outliers in {subject}:\n", outliers)
```

```python
import random

random_float = random.random()
print("Random float between 0 and 1:", random_float)

random_int = random.randint(1, 100)
print("Random integer between 1 and 100:", random_int)

random_range = random.randrange(1, 100)
print("Random integer between 1 and 99:", random_range)

random_uniform = random.uniform(1.5, 10.5)
print("Random float between 1.5 and 10.5:", random_uniform)

random_choice = random.choice(['apple', 'banana', 'cherry', 'date'])
print("Random choice from list:", random_choice)

random_sample = random.sample(range(1, 100), 5)
print("Random sample of 5 elements from range 1 to 99:", random_sample)

data_list = [1, 2, 3, 4, 5]
random.shuffle(data_list)
print("Shuffled list:", data_list)

random_normal = random.gauss(0, 1)
print("Random float from normal distribution (mean=0, std=1):", random_normal)

random_exponential = random.expovariate(1)
print("Random float from exponential distribution (lambda=1):", random_exponential)

random_beta = random.betavariate(2, 5)
print("Random float from beta distribution (alpha=2, beta=5):", random_beta)
```
