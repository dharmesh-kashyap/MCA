## Filling the Null values

```python
df2 = df.fillna(value=0)
df2
```

```python
df2.isnull().sum().sum()
```

```python
df.head(10)
```

```python
df3 = df.fillna(method='pad') #Filling the null values with previous valid value
df3.head(10)
```

```python
# bfill - Filling the null values with next values, row wise.
df4 = df.fillna(method='bfill')
df4.head(10)
```

```python
df5 = df.fillna(method='pad', axis=1)
df5.head(10)
```

```python
df6 = df.fillna({'society':'abcd', 'balcony':'defg'})
df6.head(10)
```

```python
df7 = df.fillna(value=df['balcony'].mean())
df7.head(10)
```

```python
df8 = df.fillna(value=df['balcony'].max())
df8.head(10)
```

```python
df9 = df.fillna(value=df['balcony'].min())
df9.head(10)
```

```python
df10 = df.dropna()
df10.head()
```

```python
df11 = df.dropna(how='any')
df11.head(10)
```

```python
df12 = df.replace(to_replace=np.nan, value=123456789)
df12
```

```python
df13 = df.replace('Plot  Area', 'xyz')
df13.head(10)
```
