```python
df1 = pd.read_csv('Bengaluru_House_Data.csv')
df1.head()
```

```python
df1.isnull() #Checking the presence of null value across all columns of the dataset
```

```python
df1.isnull().sum() #Count the no.of null value present in each columns of the dataset
```

```python
df2 = df1.fillna(value = 0) #Fill the null value with Zero
df2
```

```python
df2.isnull().sum() #Now all the Na value is Zero
```

```python
df3 = df1.fillna(value = 5) # Now all the Null values is replaces by 5
df3
```

```python
df4 = df1.fillna(method = 'pad') #Filling null value with previous value
df4
```

```python
df5 = df1.fillna(method = 'bfill') #Filling null value with later value
df5
```

```python
#Filling Null value with next column
df7 = df1.fillna(method = 'bfill', axis = 1)
df7
```

```python
df8 = df1.fillna({'society':'abcd','balcony':'defg'}) #Filling values in Null in diffrent columns
df8
```

```python
df9=df1.fillna(value = df1['balcony'].mean())
df9
```

```python
#Dropping thje rows thwt contain Null values
df11 = df1.dropna()
df11
```

```python
df12 = df1.dropna(how = 'any')
df12
```

```python
df13 = df1.dropna(how = 'all')
df13
```

```python
#The function has replaced all values in the data set with the value 875465
df14 = df1.replace(to_replace=np.nan,value = 875465)
df14
```

```python
#Replacing the existing value by new value
df15 = df1.replace(to_replace=3.0,value = 5.0)
df15
```
