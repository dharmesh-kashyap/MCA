```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# ======= 1. Data Loading and Preparation ======= #

# Load datasets
insurance_data = pd.read_csv('insurance.csv')
salaries_data = pd.read_csv('Salaries.csv')

# Get dummies for categorical columns in insurance data
insurance_data = pd.get_dummies(insurance_data, columns=['smoker', 'sex', 'region'], drop_first=True)

# Prepare data for SLR
X_slr = insurance_data[['smoker_yes', 'sex_male']].values  # Predictor variables
y_slr = insurance_data['charges'].values  # Target variable

# Get dummies for categorical columns in salaries data
salaries_data = pd.get_dummies(salaries_data, columns=['sex', 'rank', 'discipline'], drop_first=True)

# Prepare data for MLR
X_mlr = salaries_data[['yrs.since.phd', 'yrs.service', 'sex_Male', 'rank_Prof', 'discipline_B']].values  # Predictor variables
y_mlr = salaries_data['salary'].values  # Target variable
```

```python
# Adding a column of ones to X for the intercept term in SLR
X_slr = np.c_[np.ones(X_slr.shape[0]), X_slr]  # Add intercept

# Gradient Descent for SLR
def gradient_descent(X, y, learning_rate=0.01, iterations=1000):
    m = len(y)
    theta = np.zeros(X.shape[1])  # Initialize theta (parameters)
    cost_history = []

    for i in range(iterations):
        prediction = X.dot(theta)  # Linear prediction
        error = prediction - y     # Error term
        cost = (1/(2*m)) * np.dot(error.T, error)  # Cost function
        cost_history.append(cost)
        gradient = (1/m) * X.T.dot(error)  # Gradient calculation
        theta = theta - learning_rate * gradient  # Update theta
    return theta, cost_history

# Apply Gradient Descent for SLR
theta_slr, cost_history_slr = gradient_descent(X_slr, y_slr, learning_rate=0.001, iterations=10000)
print("Theta values for SLR:", theta_slr)
print("Final cost for SLR:", cost_history_slr[-1])
```
