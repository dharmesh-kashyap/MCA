```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

import warnings
def warn(*args, **kwargs):
    pass
warnings.warn = warn
```

```python
df = pd.read_csv(r"D:\240970058\OneDrive - Manipal Academy of Higher Education\Desktop\MLDATASETS-19AUG2024\cancer.csv.csv")
df.head()
```

```python
df.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
X = df.drop('diagnosis', axis=1)
y = df.diagnosis
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
pca = PCA(n_components=2)
principal_comps = pca.fit_transform(X_scaled)
principal_df = pd.DataFrame(principal_comps, columns=['PC 1','PC 2'])
principal_df
```

```python
print(pca.explained_variance_ratio_)
```

```python
import csv
with open('pca_cancer_results.csv', 'w', newline='') as file:
    csvWriter = csv.writer(file)
    csvWriter.writerow(principal_df.columns.tolist())
    csvWriter.writerows(principal_df.values)
```
