```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
df=pd.read_csv("../MLDATASETS-19AUG2024/diabetes.csv")
df[:5]
```

```python
x=df.iloc[:,1:5]
target=df.iloc[:,-1]
ss=StandardScaler()
x=ss.fit_transform(x)
print(x)
```

```python
lda = LDA(n_components=1)
l=lda.fit_transform(x,target)
print(l)
```

```python
plt.scatter(l,target,c=target)
```
