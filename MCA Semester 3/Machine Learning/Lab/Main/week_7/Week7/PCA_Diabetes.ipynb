```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import warnings
def warn(*args, **kwargs):
    pass
warnings.warn = warn
```

```python
df = pd.read_csv(r"D:\240970058\OneDrive - Manipal Academy of Higher Education\Desktop\MLDATASETS-19AUG2024\diabetes.csv")
df.head()
```

```python
X = df.drop('Outcome', axis=1)
y = df.Outcome
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
pca = PCA(n_components=2)

principal_comps = pca.fit_transform(X_scaled)
principal_df = pd.DataFrame(principal_comps, columns=['PC 1','PC 2'])
principal_df
```

```python
class_names = ['Outcome: 0','Outcome: 1']
for i, class_name in enumerate(class_names):
    plt.scatter(principal_comps[y == i, 0], principal_comps[y == i, 1], label=class_name)
plt.legend()
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization')
```

```python
print(pca.explained_variance_ratio_)
```

```python
import csv
with open('pca_diabetes_results.csv','w',newline='') as file:
    csvWriter = csv.writer(file)
    csvWriter.writerow(principal_df.columns.tolist())
    csvWriter.writerows(principal_df.values)
```
