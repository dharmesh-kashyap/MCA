```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
df=pd.read_csv("../MLDATASETS-19AUG2024/IRIS.csv")
df[:5]
```

```python
x=df.iloc[:,1:5]
target=df.iloc[:,-1]
enc=LabelEncoder()
target=enc.fit_transform(target)
ss=StandardScaler()
x=ss.fit_transform(x)
print(x)
```

```python
lda = LDA(n_components=2)
l=lda.fit_transform(x,target)
print(l)
```

```python
plt.scatter(l[:,0],l[:,1],c=target)
plt.xlabel("Principal 1")
plt.ylabel("Principal 2")
```
