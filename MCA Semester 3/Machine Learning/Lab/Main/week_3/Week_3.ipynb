```python
import numpy as np

# Reading matrices from user input
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

print("Enter the elements of the matrix row by row, separated by spaces:")
matrix = []
for _ in range(rows):
    while True:
        row = list(map(int, input().split()))
        if len(row) == cols:
            matrix.append(row)
            break
        else:
            print(f"Please enter exactly {cols} elements.")

matrix = np.array(matrix)
print("Matrix:")
print(matrix)

# Writing the matrix to a file
np.savetxt('matrix.txt', matrix, fmt='%d')
print("Matrix written to 'matrix.txt' file.")
```

```python
# Define two matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Matrix Addition
addition_result = matrix1 + matrix2
print("Matrix Addition Result:")
print(addition_result)

# Matrix Multiplication
multiplication_result = np.dot(matrix1, matrix2)
print("Matrix Multiplication Result:")
print(multiplication_result)
```

```python
# Define a square matrix
matrix = np.array([[1, 2], [3, 4]])

# Matrix Transpose
transpose_matrix = np.transpose(matrix)
print("Transpose of the Matrix:")
print(transpose_matrix)

# Check if the matrix is invertible
if np.linalg.det(matrix) == 0:
    print("Matrix is not invertible.")
else:
    # Matrix Inversion
    inverse_matrix = np.linalg.inv(matrix)
    print("Inverse of the Matrix:")
    print(inverse_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 7, 11])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot the data and the line of best fit
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Line of best fit')
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Display the linear regression equation and parameters
intercept = model.intercept_
slope = model.coef_[0]
print(f"Linear Regression Equation: y = {slope:.2f}x + {intercept:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"Slope: {slope:.2f}")
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load the California housing dataset
data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['MedHouseVal'] = data.target
print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Prepare the data
X = df[['AveRooms']].values
y = df['MedHouseVal'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot the data and the line of best fit
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Line of best fit')
plt.title("Linear Regression on California Housing Dataset")
plt.xlabel("Average Number of Rooms")
plt.ylabel("Median House Value")
plt.legend()
plt.show()

# Display the linear regression equation and parameters
intercept = model.intercept_
slope = model.coef_[0]
print(f"Linear Regression Equation: MedHouseVal = {slope:.2f} * AveRooms + {intercept:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"Slope: {slope:.2f}")
```

```python
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
data = load_diabetes()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
print(df.head())
```

```python
# Prepare the data
X = df[['bmi']].values
y = df['target'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot the data and the line of best fit
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Line of best fit')
plt.title("Linear Regression on Diabetes Dataset")
plt.xlabel("Body Mass Index (BMI)")
plt.ylabel("Disease Progression")
plt.legend()
plt.show()

# Display the linear regression equation and parameters
intercept = model.intercept_
slope = model.coef_[0]
print(f"Linear Regression Equation: target = {slope:.2f} * bmi + {intercept:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"Slope: {slope:.2f}")
```
