```python
#program1
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(0, 1, 100)  # Normally distributed data for histogram and box plot
categories = ['A', 'B', 'C', 'D']    # Categories for bar and pie charts
values = [15, 30, 45, 10]            # Values corresponding to categories
x = np.random.rand(100)              # Random x values for scatter plot
y = np.random.rand(100)              # Random y values for scatter plot
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='green', alpha=0.7)
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow'])
plt.title('Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='purple', alpha=0.7)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

```python
#program2

import numpy as np

def create_matrix(m, n):
    print(f"Enter the elements for a {m}x{n} matrix row by row:")
    matrix = []
    for i in range(m):
        row = []
        for j in range(n):
            element = np.random.randint(10)
            row.append(element)
        matrix.append(row)
    return np.array(matrix)

M = int(input("Enter the number of rows (M): "))
N = int(input("Enter the number of columns (N): "))

matrix = create_matrix(M, N)
print(f"{M}x{N} Matrix:")
print(matrix)
```

```python
print("Matrix Addition:")
matrix_addition = matrix+matrix
print(np.array(matrix_addition))

print("Matrix Multiplication:")
matrix_multiplication = np.dot(matrix, matrix)
print(np.array(matrix_multiplication))
```

```python
inverted_matrix = np.linalg.inv(matrix)
print("Inverted Matrix:")
print(inverted_matrix)

transposed_matrix  = matrix.T
print("Transposed Matrix:")
print(transposed_matrix)
```

```python
#program3

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# Sample data
l1 = [71, 89, 64, 74, 87, 93, 79, 81, 75, 72, 70, 66, 81, 78, 110]  # Waist_Cm
l2 = [51, 66, 45, 57, 64, 75, 61, 62, 55, 53, 48, 46, 63, 60, 100]  # Weight_kg

# Create DataFrame
data = {'Waist_Cm': l1, 'Weight_kg': l2}
df = pd.DataFrame(data)

# Define the features (X) and the target variable (y)
X = df[['Waist_Cm']]  # Independent variable
y = df['Weight_kg']   # Dependent variable

# Create a Simple Linear Regression model
model = linear_model.LinearRegression()

# Fit the model
model.fit(X, y)

# Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Display linear regression equation
print(f'Linear Regression Equation: Weight_kg = {slope:.2f} * Waist_Cm + {intercept:.2f}')

# Predict values for plotting the line of best fit
y_pred = model.predict(X)

# Plotting the data points and the best fit line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Line of Best Fit')
plt.xlabel('Waist (cm)')
plt.ylabel('Weight (kg)')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid()
plt.show()
```

```python
#program4
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load mtcars dataset from a URL
mtcars_df = pd.read_csv('mtcars.csv')

# Linear Regression on mtcars dataset
X_mtcars = mtcars_df[['wt']]  # Independent variable: Weight
y_mtcars = mtcars_df['mpg']    # Dependent variable: Miles Per Gallon

# Create and fit the model
model_mtcars = LinearRegression()
model_mtcars.fit(X_mtcars, y_mtcars)

# Predict values for the regression line
y_pred_mtcars = model_mtcars.predict(X_mtcars)

# Plotting for mtcars
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_mtcars, y_mtcars, color='blue', label='Data Points')
plt.plot(X_mtcars, y_pred_mtcars, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression on mtcars dataset')
plt.xlabel('Weight of Car (wt)')
plt.ylabel('Miles Per Gallon (mpg)')
plt.legend()
plt.grid()

# Load abalone dataset
abalone_df = pd.read_csv('abalone.csv', 
                          header=None, names=['Sex', 'Length', 'Diameter', 'Height', 'WholeWeight', 'ShuckedWeight', 'VisceraWeight', 'ShellWeight', 'Rings'])

# Linear Regression on abalone dataset
X_abalone = abalone_df[['Length']]  # Independent variable: Length
y_abalone = abalone_df['Rings']      # Dependent variable: Rings

# Create and fit the model
model_abalone = LinearRegression()
model_abalone.fit(X_abalone, y_abalone)

# Predict values for the regression line
y_pred_abalone = model_abalone.predict(X_abalone)

# Plotting for abalone
plt.subplot(1, 2, 2)
plt.scatter(X_abalone, y_abalone, color='green', label='Data Points')
plt.plot(X_abalone, y_pred_abalone, color='orange', linewidth=2, label='Regression Line')
plt.title('Linear Regression on abalone dataset')
plt.xlabel('Length (cm)')
plt.ylabel('Number of Rings')
plt.legend()
plt.grid() 

plt.tight_layout()
plt.show()

# Output parameters for mtcars
print(f'mtcars Linear Regression Equation: mpg = {model_mtcars.coef_[0]:.2f} * wt + {model_mtcars.intercept_:.2f}')
# Output parameters for abalone
print(f'Abalone Linear Regression Equation: Rings = {model_abalone.coef_[0]:.2f} * Length + {model_abalone.intercept_:.2f}')
```
