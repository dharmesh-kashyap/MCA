```python
from sklearn import datasets
import numpy as np
wine = datasets.load_wine()

print(wine.feature_names)
print(wine.target_names)
print(wine.data[0:5])
print((wine.target).size)
print(len(wine.target))
print(wine.data.shape)
print(wine.target.shape)

indices = np.arange(len(wine.data))


from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report, confusion_matrix
from tabulate import tabulate
test_sizes = [0.3,0.4,0.5]
for ts in test_sizes:
    print(f"\n\n\nFor test size {ts}:")
    print("---------------------------------------------------------------------------------------------------")
    x_train, x_test, y_train, y_test, train_indices, test_indices = train_test_split(
        wine.data, wine.target, indices, test_size=ts, random_state=55
    )

    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train,y_train)

    predicted=model.predict(x_test)
    accuracy = accuracy_score(y_test,predicted)
    headers = ['Test_percentage','Actual', 'Predicted']
    table = [[ts,wine.target_names[y_test[i]], wine.target_names[predicted[i]]] for i in range(len(x_test))]
    limited_table = table[:10]
    print(f"\nindices of the train data: {train_indices}")
    print(f"\nindices of the test data: {test_indices}")
    print(f"\n\npredicted:{predicted}")
    print(f"\nConfusion Matrix: \n{confusion_matrix(y_test,predicted)}")
    print(f"\nscore:{model.score(x_train,y_train)}")
    print(f"\naccuracy:{accuracy}")
    print(f"\nClassification Report:\n{classification_report(y_test, predicted, target_names=wine.target_names)}")
    print(tabulate(table, headers=headers, tablefmt="grid"))
```
