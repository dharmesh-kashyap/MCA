```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv('income.data.csv')
x = df[['income']].values
y = df['happiness'].values

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)

model = LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

#print(y_pred)
print(f"Line equation : y = {model.coef_[0]:.2f}x+{model.intercept_:.2f}")
rdf = pd.DataFrame({'x_test(income)':x_test.flatten(),
                   'y_test(actual happiness)':y_test,
                   'y_pred(predicted happiness)':y_pred
})
print("\n",rdf,"\n\n")

plt.scatter(x_test,y_test, color='skyblue',label='Actual')
plt.plot(x_test ,y_pred , color='red' ,linewidth=4 ,label='regression line')
plt.xlabel('Income')
plt.ylabel('Happiness')
plt.title('Income vs Happiness')
```

```python
import numpy as np

def read_matrix(prompt):
    print(prompt)
    rows = int(input("Rows: "))
    cols = int(input("Cols: "))
    matrix = [list(map(float, input(f"Row {i + 1}: ").split())) for i in range(rows)]
    if any(len(row) != cols for row in matrix):
        raise ValueError("Column count mismatch.")
    return np.array(matrix)

def print_matrix(matrix, name):
    print(f"{name}:")
    print(matrix)

def matrix_operations(matrix1, matrix2):
    if matrix1.shape != matrix2.shape:
        raise ValueError("Dimension mismatch for addition.")
    if matrix1.shape[1] != matrix2.shape[0]:
        raise ValueError("Dimension mismatch for multiplication.")
    
    addition = matrix1 + matrix2
    multiplication = np.dot(matrix1, matrix2)
    transpose1 = matrix1.T
    transpose2 = matrix2.T
    inversion1 = np.linalg.inv(matrix1) if matrix1.shape[0] == matrix1.shape[1] else "Not square"
    inversion2 = np.linalg.inv(matrix2) if matrix2.shape[0] == matrix2.shape[1] else "Not square"
    
    return addition, multiplication, transpose1, transpose2, inversion1, inversion2

if __name__ == "__main__":
    matrix1 = read_matrix("Input for Matrix 1")
    matrix2 = read_matrix("Input for Matrix 2")
    
    try:
        addition, multiplication, transpose1, transpose2, inversion1, inversion2 = matrix_operations(matrix1, matrix2)
        
        print_matrix(matrix1, "Matrix 1")
        print_matrix(matrix2, "Matrix 2")
        print_matrix(addition, "Matrix Addition")
        print_matrix(multiplication, "Matrix Multiplication")
        print_matrix(transpose1, "Matrix 1 Transpose")
        print_matrix(transpose2, "Matrix 2 Transpose")
        print_matrix(inversion1, "Matrix 1 Inversion")
        print_matrix(inversion2, "Matrix 2 Inversion")
        
    except ValueError as e:
        print(e)
```
