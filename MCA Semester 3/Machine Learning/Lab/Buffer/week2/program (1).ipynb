```python
#program3
import numpy as np

# Part a) Create an M x N Matrix using NumPy arrays
def create_matrix():
    M = int(input("Enter the number of rows (M): "))
    N = int(input("Enter the number of columns (N): "))
    print("Enter the elements of a {M}x{N} matrix row-wise:")
    matrix = np.zeros((M, N))
    for i in range(M):
        row = list(map(float, input("Enter row {i+1}: ").split()))
        if len(row) != N:
            print("Error: Expected {N} columns, got {len(row)}. Please try again.")
            return create_matrix()
        matrix[i] = row
    return matrix

# Part b) Matrix Addition and Multiplication using Python lists and arrays
def matrix_addition(A, B):
    # Using Python lists
    result = [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
    return result

def matrix_multiplication(A, B):
    # Using Python lists
    result = [[sum(A[i][k] * B[k][j] for k in range(len(B))) for j in range(len(B[0]))] for i in range(len(A))]
    return result

# Part c) Matrix Inversion using NumPy and Python lists
def matrix_inversion(matrix):
    # Using NumPy
    try:
        inv_matrix = np.linalg.inv(matrix)
        return inv_matrix
    except np.linalg.LinAlgError:
        print("Matrix inversion is not possible (matrix might be singular).")
        return None

def matrix_inversion_lists(matrix):
    # Using Python lists
    n = len(matrix)
    inverse = [[0.0] * n for _ in range(n)]
    identity = [[float(i == j) for i in range(n)] for j in range(n)]
    
    for i in range(n):
        for j in range(n):
            inverse[i][j] = matrix[i][j]
    
    for i in range(n):
        diag = inverse[i][i]
        if diag == 0:
            print("Matrix inversion is not possible (matrix might be singular).")
            return None
        
        for j in range(n):
            inverse[i][j] /= diag
            identity[i][j] /= diag
            
        for k in range(n):
            if i != k:
                factor = inverse[k][i]
                for j in range(n):
                    inverse[k][j] -= factor * inverse[i][j]
                    identity[k][j] -= factor * identity[i][j]
    
    return identity

# Main function to demonstrate the operations
def main():
    print("Creating matrix A:")
    A = create_matrix()
    print("Creating matrix B:")
    B = create_matrix()

    # Check if the matrices can be added and multiplied
    if A.shape != B.shape:
        print("Matrix addition is not possible (different dimensions).")
    else:
        addition_result = matrix_addition(A, B)
        print("\nMatrix Addition Result:")
        print(np.array(addition_result))

    if A.shape[1] != B.shape[0]:
        print("Matrix multiplication is not possible (inner dimensions do not match).")
    else:
        multiplication_result = matrix_multiplication(A, B)
        print("\nMatrix Multiplication Result:")
        print(np.array(multiplication_result))
    
    print("\nMatrix Inversion using NumPy:")
    inv_A = matrix_inversion(A)
    if inv_A is not None:
        print(inv_A)
    
    print("\nMatrix Inversion using Python lists:")
    inv_A_lists = matrix_inversion_lists(A.tolist())
    if inv_A_lists is not None:
        print(np.array(inv_A_lists))

if __name__ == "__main__":
    main()
```

```python
#program6

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Create Sample Data
data = {
    'SNO': range(1, 11),
    'REGNO': [f'R00{i}' for i in range(1, 11)],
    'NAME': ['Student1', 'Student2', 'Student3', 'Student4', 'Student5', 
             'Student6', 'Student7', 'Student8', 'Student9', 'Student10'],
    'M1': [85, 78, 92, 67, 74, 88, 90, 76, 84, 70],
    'M2': [75, 82, 89, 95, 80, 73, 85, 87, 77, 88],
    'M3': [91, 85, 78, 82, 89, 84, 79, 80, 86, 92],
    'M4': [88, 79, 95, 90, 85, 82, 81, 94, 89, 83],
    'M5': [80, 88, 86, 75, 82, 77, 90, 84, 78, 85],
}

# Create DataFrame
df = pd.DataFrame(data)

# Step 4: Save to Excel
df.to_excel('student_data.xlsx', index=False)

# Step 5: Compute Statistical Measures
mean_values = df[['M1', 'M2', 'M3', 'M4', 'M5']].mean()
median_values = df[['M1', 'M2', 'M3', 'M4', 'M5']].median()
mode_values = df[['M1', 'M2', 'M3', 'M4', 'M5']].mode().iloc[0]
summary = df[['M1', 'M2', 'M3', 'M4', 'M5']].describe()

print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
print("\n5 Number Summary:\n", summary)

# Step 6: Box Plot for All Marks
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['M1', 'M2', 'M3', 'M4', 'M5']])
plt.title('Box Plot for Marks in All Subjects')
plt.show()

# Step 7: Various Plots for Each Subject
for col in ['M1', 'M2', 'M3', 'M4', 'M5']:
    
    sns.histplot(df[col])
    plt.title(f'Histogram for {col}')
    plt.show()

    
    sns.scatterplot(x='SNO', y=col, data=df)
    plt.title(f'Scatter Plot for {col}')
    plt.show()

# Step 8: Identify and Display Outliers
def find_outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return data[(data < lower_bound) | (data > upper_bound)]

for col in ['M1', 'M2', 'M3', 'M4', 'M5']:
    outliers = find_outliers(df[col])
    if not outliers.empty:
        print(f"Outliers in {col}: {outliers.values}")
    else:
        print(f"No outliers in {col}.")
```
