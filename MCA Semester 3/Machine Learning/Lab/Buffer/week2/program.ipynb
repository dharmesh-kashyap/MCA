```python
import numpy as np

array = np.linspace(0, 10, num=5)
print("Array: ", array)

array, step = np.linspace(0, 10, num=5, retstep=True)
print("Array with step: ", array)
print("Step size: ", step)

array = np.linspace(0, 10, num=5, endpoint=False)
print("Array without endpoint:", array)

import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 10)
y = np.random.rand(9) * 10


plt.scatter(x, y, color='blue', marker='o')
plt.title('Simple Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Annotating all points on the plot with their coordinates
for xi, yi in zip(x, y):
    plt.annotate(f"({xi}, {yi:.2f})", (xi, yi), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.plot(data)
plt.title('Random Data Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, alpha=0.7, color='green')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

plt.bar(categories, values, color='cyan')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.randn(100) for _ in range(5)]

plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```
