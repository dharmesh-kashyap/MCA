```python
import numpy as np
print("enter matrix1 value in a single line (separated by space): ")
m1 = list(map(int, input().split()))

print("enter matrix2 value in a single line (separated by space): ")
m2 = list(map(int, input().split()))

print("Dimension of m1: ", np.shape(m1))
print("Dimension of m2: ", np.shape(m2))

mat1 = np.array(m1).reshape(2,3)
mat2 = np.array(m2).reshape(3,2)
print(np.shape(mat1))
print(np.shape(mat2))

mul1 = np.matmul(mat1, mat2)
print(mul1)
print("Dimension of mul1: ", np.shape(mul1))
x2 = np.matmul(mul1, mul1.T)
print(x2)
print(np.shape(x2))

invx2 = np.linalg.inv(x2)
print(invx2)
```

```python
import math
import numpy as np

x = np.arange(0, math.pi*2, 0.05)
print(x)
print(len(x))
y = np.sin(x)
print(y)

plt.scatter(x,y)
plt.xlabel("angle")
plt.ylabel("sine of x")
plt.title("Sine Curve")
plt.show()
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

data_dict = {'CSE':33, 'ECE':28, 'EEE':30}

courses = list(data_dict.keys())
values = list(data_dict.values())

fig = plt.figure(figsize=(10,5))

plt.bar(courses, values, color='green', width=.5)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
